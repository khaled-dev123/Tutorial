#Pandas: Reading Data and Analyzing DataFrames
##1. Reading CSV and JSON Files
###1.1 Read CSV Files

```python
import pandas as pd

# Load CSV file into a DataFrame
df_csv = pd.read_csv('data.csv')

# Print entire DataFrame
print(df_csv.to_string())
```

###1.2 Read JSON Files

```python
# Load JSON file into a DataFrame
df_json = pd.read_json('data.json')

print(df_json.to_string())
```

##2. Viewing the Data
###2.1 Using head()

The head() method returns the top rows of the DataFrame.
By default, it returns 5 rows, but you can specify a number:

```python
# First 10 rows
print(df_csv.head(10))

# First 5 rows (default)
print(df_csv.head())
```

###2.2 Using tail()

The tail() method returns the last rows of the DataFrame:

```python
# Last 5 rows
print(df_csv.tail())
```

##3. Get Info About the Data

The info() method provides a summary of the DataFrame:

```python
print(df_csv.info())
```

Example Output:

##4. Handling Null Values

Empty or null values can affect data analysis. You may want to clean your data before analysis.

```python
# Remove rows with any null values
df_clean = df_csv.dropna()

# Or fill null values with 0
df_filled = df_csv.fillna(0)
```
