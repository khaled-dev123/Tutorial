#Pandas DataFrame Tutorial

##1. What is a DataFrame?

A Pandas DataFrame is a 2-dimensional data structure, like a table with rows and columns.
It is similar to a 2D array, but more flexible because it can hold different types of data in each column.
Example: Creating a Simple DataFrame

```python
import pandas as pd

# Create a dictionary with data
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

# Load data into a DataFrame
df = pd.DataFrame(data)

print(df)
```

##2. Locate Rows

Pandas uses the loc attribute to access one or more specific rows.

Return a Single Row

```python
print(df.loc[0])
```

###Return Multiple Rows

```python
print(df.loc[[0, 1]])
```

##3. Named Indexes

You can assign custom row names using the index argument.

```python
df = pd.DataFrame(data, index=["day1", "day2", "day3"])
print(df)
```

###Access Rows by Name

```python
print(df.loc["day2"])
```

##4. Load Files into a DataFrame

If your data is stored in a file (like CSV), Pandas can load it directly into a DataFrame.

```python
import pandas as pd

# Load data from a CSV file
df = pd.read_csv('data.csv')

print(df)
```

###5. Change Maximum Rows Displayed

You can increase the number of rows shown:

```python
import pandas as pd

# Set maximum rows to display
pd.options.display.max_rows = 9999

# Load the CSV
df = pd.read_csv('data.csv')

# Print the entire DataFrame
print(df)
```
