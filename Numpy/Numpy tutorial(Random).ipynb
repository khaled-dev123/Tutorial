#Random Numbers in NumPy

##üé≤ Random Numbers in NumPy
###üîπ What is a Random Number?

A random number does not mean it changes every time.
It means a number that cannot be predicted logically.

###üîπ Pseudo-Random vs True Random

Computers run on algorithms (fixed instructions).

If a computer generates numbers with an algorithm, they are pseudo-random (not truly random, but look random).

‚úÖ Pseudo-Random Numbers ‚Üí Generated using algorithms.                              
‚úÖ True Random Numbers ‚Üí Come from outside sources like mouse movements, keystrokes, or natural events.

üëâ We usually don‚Äôt need true randomness, unless for security (encryption) or games of chance.
In NumPy, we work with pseudo-random numbers.

###üîπ Generate Random Numbers in NumPy

NumPy provides the random module for generating random numbers.

‚úÖ Example 1: Random Integer

```python
from numpy import random

x = random.randint(100)  # Random integer from 0 to 99
print(x)
```

‚úÖ Example 2: Random Float

```python
from numpy import random

x = random.rand()  # Random float between 0 and 1
print(x)
```

###üîπ Generate Random Arrays

We can also generate entire arrays of random values.                                           
‚úÖ Example 3: 1-D Array of Random Integers

```python
from numpy import random

x = random.randint(100, size=5)
print(x)
```

‚úÖ Example 4: 2-D Array of Random Integers

```python
from numpy import random

x = random.randint(100, size=(3, 5))  # 3 rows, 5 columns
print(x)
```

‚úÖ Example 5: 1-D Array of Random Floats

```python
from numpy import random

x = random.rand(5)
print(x)
```

‚úÖ Example 6: 2-D Array of Random Floats

```python
from numpy import random

x = random.rand(3, 5)
print(x)
```

###üîπ Random Selection from an Array

The choice() method picks random values from a given array.

‚úÖ Example 7: Pick One Random Value

```python
from numpy import random

x = random.choice([3, 5, 7, 9])
print(x)
```

‚úÖ Example 8: Pick Multiple Values

```python
from numpy import random

x = random.choice([3, 5, 7, 9], size=(3, 5))
print(x)
```

## Data Distribution in NumPy
### What is Data Distribution?

A data distribution is a list of all possible values and how often each value occurs.

It is a fundamental concept in statistics and data science.

NumPy allows us to generate random data distributions using the random module.

### Random Distribution

A random distribution is a set of random numbers that follow a certain probability density function (PDF).

Probability Density Function (PDF):
A function that describes the probability of different values in a dataset.


### Generate Random Numbers with Custom Probabilities

NumPy‚Äôs choice() method allows you to:

üëâSelect values from a given list.

üëâDefine the probability of each value.

üëâ The probabilities must add up to 1.

üìù Example 1: 1-D Array with Probabilities

```python
from numpy import random

# Generate 100 values from [3, 5, 7, 9]
# With given probabilities
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))

print(x)
```

‚úÖ Explanation:

3 ‚Üí 10% chance

5 ‚Üí 30% chance

7 ‚Üí 60% chance

9 ‚Üí 0% chance (will never appear

### Example 2: 2-D Array with Probabilities

```python
from numpy import random

# Generate a 2-D array with 3 rows and 5 columns
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))

print(x)
```

‚úÖ Output will be a matrix where each element is randomly chosen from [3, 5, 7, 9] with the given probabilities.

‚ö° Key Notes:

The sum of probabilities must equal 1.0.

Setting probability 0 means the value will never appear.

You can generate arrays of any shape with the size parameter.

## Random Permutations in NumPy
### What is a Permutation?

A permutation is an arrangement of elements in a particular order.

In NumPy, we can create permutations (shuffled orders of arrays) using the random module.

There are two main functions:

random.shuffle() ‚Üí Shuffles the array in-place (modifies the original array).

random.permutation() ‚Üí Returns a new shuffled array without changing the original.

### Example 1: Shuffle an Array (in-place)

```python
from numpy import random
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
random.shuffle(arr)

print(arr)
```



### Example 2: Generate a Random Permutation

```python
from numpy import random
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
perm = random.permutation(arr)

print("Original:", arr)
print("Permutation:", perm)
```

### Example 3: Permutation of Integers

You can also generate a permutation of integers without providing an array:

```python
from numpy import random

x = random.permutation(10)  # numbers from 0 to 9

print(x)
```

‚ö° Key Differences:

shuffle() ‚Üí changes the original array (in-place).

permutation() ‚Üí creates a new shuffled array (doesn‚Äôt touch original).

## Normal (Gaussian) Distribution in NumPy
### What is Normal Distribution?

The Normal Distribution is one of the most used probability distributions in statistics.

It is also called the Gaussian Distribution (named after Carl Friedrich Gauss).

Data tends to cluster around a central value (the mean) with no bias to the left or right.

The graph looks like a bell curve.

### Formula of Normal Distribution

The probability density function of a normal distribution is:

f(x) = (1 / (œÉ * sqrt(2œÄ))) * e^(-0.5 * ((x - Œº) / œÉ)^2)


Where:

Œº (mu) = mean (center of the distribution).

œÉ (sigma) = standard deviation (spread of the data).

üîπ Generate Normal Distribution in NumPy

NumPy provides the function:

```python
random.normal(loc, scale, size)
```

loc ‚Üí mean (center).

scale ‚Üí standard deviation (spread).

size ‚Üí shape of the returned array.


### Example 1: Generate 10 Random Numbers

```python
from numpy import random

x = random.normal(loc=0, scale=1, size=10)
print(x)
```

###Example 1: Generate 10 Random Numbers

```python
from numpy import random

x = random.normal(loc=0, scale=1, size=10)
print(x)
```

### Example 2: 2D Normal Distribution

```python
from numpy import random

x = random.normal(loc=5, scale=2, size=(3, 5))
print(x)
```

### Example 3: Visualization (Bell Curve)

```python
from numpy import random
import matplotlib.pyplot as plt

x = random.normal(loc=0, scale=1, size=1000)

plt.hist(x, bins=30, density=True, color='skyblue', edgecolor='black')
plt.title("Normal Distribution (mean=0, std=1)")
plt.show()
```

‚ö° Key Notes:

Most data in nature and real-world scenarios (like heights, exam scores, errors in measurement) follow a normal distribution.

Mean defines the center.

Standard deviation defines the spread (narrow vs wide curve).

## Binomial Distribution in NumPy
### What is Binomial Distribution?

The Binomial Distribution describes the number of successes in a fixed number of independent yes/no experiments (also called Bernoulli trials).

Each trial has only two outcomes: success or failure.

Examples in real life:

Tossing a coin 10 times ‚Üí How many heads?

Rolling a dice 6 times ‚Üí How many times 4 appears?
### Formula of Binomial Distribution

The probability of getting exactly k successes in n trials is:

P(X = k) = (nCk) * (p^k) * (q^(n-k))

Where:

n = number of trials

k = number of successes

p = probability of success in one trial

q = 1 - p = probability of failure

### NumPy Implementation

NumPy provides the function:

random.binomial(n, p, size)


n ‚Üí number of trials

p ‚Üí probability of success in each trial

size ‚Üí shape of returned array

### Example 1: Tossing a Coin


```python
from numpy import random

x = random.binomial(n=10, p=0.5, size=10)
print(x)
```

### Example 2: Rolling a Dice

Suppose the probability of rolling a 6 is 1/6 ‚âà 0.166.

```python
from numpy import random

x = random.binomial(n=6, p=1/6, size=10)
print(x)
```

### Example 3: Visualization

```python
from numpy import random
import matplotlib.pyplot as plt

x = random.binomial(n=10, p=0.5, size=1000)

plt.hist(x, bins=10, color='lightgreen', edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

‚ö° Key Difference from Normal Distribution:

Binomial is discrete (counts successes).

Normal is continuous (values around a mean).

Uniform Distribution in NumPy
### What is Uniform Distribution?

The Uniform Distribution is a probability distribution where all outcomes are equally likely.

Every number in the interval has the same chance of being chosen.

### Example:

Rolling a fair dice ‚Üí each face (1‚Äì6) has equal probability (1/6).

Picking a random number between 0 and 1 ‚Üí every number has the same probability.

### Types of Uniform Distribution

Discrete Uniform Distribution ‚Üí Outcomes are whole numbers (e.g., dice rolls).

Continuous Uniform Distribution ‚Üí Outcomes can take any value in an interval (e.g., between 0 and 1).

### Formula for Continuous Uniform Distribution

For a uniform distribution between a and b:

f(x) = 1 / (b - a),    for a ‚â§ x ‚â§ b


Meaning: probability density is constant.

### NumPy Implementation

NumPy provides:

random.uniform(low, high, size)


low ‚Üí start of interval.

high ‚Üí end of interval.

size ‚Üí shape of returned array.

### Example 1: Random Floats

```python
from numpy import random

x = random.uniform(0, 1, 10)
print(x)
```

### Example 2: Between 2 and 6

```python
from numpy import random

x = random.uniform(2, 6, size=(3, 5))
print(x)
```

### Example 3: Visualization

```python
from numpy import random
import matplotlib.pyplot as plt

x = random.uniform(0, 10, 1000)

plt.hist(x, bins=20, color='skyblue', edgecolor='black')
plt.title("Uniform Distribution (0 to 10)")
plt.show()
```
